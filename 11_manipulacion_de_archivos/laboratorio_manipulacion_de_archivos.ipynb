***El Algoritmo k-Nearest Neighbors (kNN)***

El algoritmo **k-Nearest Neighbors (kNN)** es un m√©todo de **clasificaci√≥n** simple, basado en la idea de que los objetos con atributos similares est√°n cerca en el espacio de datos.

*Funcionamiento en la Clasificaci√≥n*

Para determinar la especie de una flor desconocida (el punto de consulta):

1.  **Medici√≥n de Distancia:** Se calcula la **distancia** (t√≠picamente **Euclidiana**) entre el arreglo de atributos (un arreglo de 4 elementos, uno por cada medici√≥n de la planta) de la nueva flor y *cada* registro de flor en el dataset. 

La f√≥rmula para la **Distancia Euclidiana** entre un punto $P=(p_1, p_2)$ y un punto $Q=(q_1, q_2)$ es:

$$
d(P, Q) = \sqrt{(p_1 - q_1)^2 + (p_2 - q_2)^2}
$$

Ahora, si el P y Q tienen m√°s de dos dimensiones (una nueva flor por clasificar $p$ que es un vector de 4 elementos, y un registro de flor $q$ en el dataset tiene 4 elementos tambi√©n), se puede calcular la distancia entre ambos como:
    $$
    d(\mathbf{P}, \mathbf{Q}) = \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2}
    $$

2.  **Selecci√≥n de Vecinos ($k$):** Se identifican los **$k$ registros de flores m√°s cercanos** (aquellos con la menor distancia).
3.  **Votaci√≥n por Mayor√≠a:** La nueva flor hereda la **especie m√°s frecuente** (la etiqueta m√°s votada) entre esos $k$ vecinos.

El valor de $k$ (el n√∫mero de vecinos) es un par√°metro del modelo.
