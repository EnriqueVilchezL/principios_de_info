La máquina de Atwood consiste en dos masas, `m1` y `m2`, conectadas por una cuerda inelástica de masa despreciable con una polea ideal de masa despreciable. Cuando `m1` = `m2`, la máquina está en equilibrio neutral sin importar la posición de los pesos. En este caso la aceleración es 0.

Cuando `m1` > `m2` ambas masas experimentan una aceleración uniforme.

Usando la segunda Ley de Newton del movimiento se puede obtener una ecuación para la aceleración del sistema de Atwood:

$$
a = g\frac{(m_1 - m_2)}{m_1 + m_2}
$$

Donde `a` es la aceleración uniforme, `g` es la gravedad, `m1` es la masa del objeto 1 y `m2` es la masa del objeto 2.
*Para efectos del programa puede suponer que m1 es mayor o igual que m2.*

Adicionalmente, la tensión de la cuerda se calcula con la siguiente fórmula:

$$
T = g\frac{2 m_1 m_2}{m_1 + m_2}
$$

La gravedad g es igual a 9.81 $metros / seg^2$, o lo que es lo mismo 981 $cm / seg^2$

El programa debe:
1. Leer 2 valores correspondientes a las masas de los objetos en la polea (en kilogramos)
2. Realizar los cálculos necesarios para encontrar la aceleración y la tensión
3. Imprimir el resultado de la aceleración en metros / seg2
4. Imprimir el resultado de la tensión en Newtons ($metros * kg / seg^2$) y en dinas ($cm * gr / seg^2$)
5. Incluir comentarios explicativos (#) para mayor legibilidad del código

Ejemplos:
| m1 (en Kg) | m2 (en Kg) | a (aceleración)     | T (tensión)                          |

| 1          | 1          | 0 mts / seg²        | 9.81 Newtons, equivalente a 981000.0 dinas  |
| 1          | 0.5        | 3.27 mts / seg²     | 6.54 Newtons, equivalente a 654000.0 dinas  |
