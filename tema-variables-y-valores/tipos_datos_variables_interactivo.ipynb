## ğŸŒŠ SecciÃ³n 3: Trabajando con Flotantes (`float`)

Los **flotantes** representan nÃºmeros reales con **precisiÃ³n finita** - pueden tener parte decimal.

### CaracterÃ­sticas principales:

- RepresentaciÃ³n de punto flotante IEEE 754

- Medidas de sensores (voltaje, temperatura, presiÃ³n)
- Constantes fÃ­sicas (Ï€, e, constante de Planck)
- CÃ¡lculos de precisiÃ³n (coordenadas GPS)

````python
# Experimentando con flotantes
print("ğŸŒŠ TRABAJANDO CON FLOTANTES")
print("=" * 30)

# Ejemplos bÃ¡sicos
voltaje = 4.95
temperatura = 23.7
pi_aproximado = 3.14159
constante_planck = 6.626e-34  # NotaciÃ³n cientÃ­fica

print(f"Voltaje del sensor: {voltaje} V")
print(f"Temperatura ambiente: {temperatura} Â°C")  
print(f"Pi aproximado: {pi_aproximado}")
print(f"Constante de Planck: {constante_planck} Jâ‹…s")

# Verificar tipos
print(f"\nTipo de voltaje: {type(voltaje)}")
print(f"Tipo de constante_planck: {type(constante_planck)}")

# Operaciones matemÃ¡ticas
print(f"\nOperaciones con flotantes:")
print(f"Temperatura en Fahrenheit: {temperatura * 9/5 + 32:.1f} Â°F")
print(f"Ãrea de cÃ­rculo (r=2.5): {pi_aproximado * 2.5**2:.2f} mÂ²")
print(f"Voltaje RMS: {voltaje / 1.414:.3f} V")

# DemostraciÃ³n de precisiÃ³n
print(f"\nPrecisiÃ³n de flotantes:")
print(f"0.1 + 0.2 = {0.1 + 0.2}")  # Â¡Sorpresa!
print(f"Â¿Es 0.1 + 0.2 == 0.3? {0.1 + 0.2 == 0.3}")
print("ğŸ’¡ Esto demuestra la limitaciÃ³n de precisiÃ³n de los flotantes")
````

## âœ…âŒ SecciÃ³n 4: Trabajando con Booleanos (`bool`)

Los **booleanos** son el tipo mÃ¡s simple: solo pueden ser `True` o `False`.

### CaracterÃ­sticas principales:
- Solo dos valores posibles: `True` y `False`
- Internamente, `True` es `1` y `False` es `0`
- Resultado de operaciones de comparaciÃ³n
- Fundamentales para la lÃ³gica de control

### Ejemplos de uso en ingenierÃ­a:
- Estados de sensores (activo/inactivo)
- Condiciones de alarma (normal/crÃ­tico)
- Flags de sistema (encendido/apagado)
- ValidaciÃ³n de datos (vÃ¡lido/invÃ¡lido)

````python
# Experimentando con booleanos
print("âœ…âŒ TRABAJANDO CON BOOLEANOS")
print("=" * 30)

# Valores bÃ¡sicos
verdadero = True
falso = False

print(f"Valor verdadero: {verdadero}")
print(f"Valor falso: {falso}")
print(f"Tipo de verdadero: {type(verdadero)}")

# Booleanos como nÃºmeros
print(f"\nBooleanos como nÃºmeros:")
print(f"True como nÃºmero: {True + 0}")
print(f"False como nÃºmero: {False + 0}")
print(f"Suma True + True: {True + True}")

# Comparaciones que generan booleanos
temperatura_actual = 95.5
temperatura_maxima = 90.0

print(f"\nğŸŒ¡ï¸ Sistema de Monitoreo de Temperatura:")
print(f"Temperatura actual: {temperatura_actual}Â°C")
print(f"Temperatura mÃ¡xima: {temperatura_maxima}Â°C")

# Operaciones de comparaciÃ³n
alarma_activada = temperatura_actual > temperatura_maxima
temperatura_normal = temperatura_actual <= temperatura_maxima
temperatura_critica = temperatura_actual > 100.0

print(f"\nEstados del sistema:")
print(f"Â¿Alarma activada? {alarma_activada}")
print(f"Â¿Temperatura normal? {temperatura_normal}")
print(f"Â¿Temperatura crÃ­tica? {temperatura_critica}")

# Operadores lÃ³gicos
sistema_seguro = temperatura_normal and not temperatura_critica
print(f"Â¿Sistema seguro? {sistema_seguro}")
````

## ğŸ·ï¸ SecciÃ³n 6: AsignaciÃ³n de Variables y Nomenclatura

### Diferencia clave: MatemÃ¡ticas vs. ComputaciÃ³n

**En MatemÃ¡ticas:** Una variable es un sÃ­mbolo para un valor **desconocido**  
- Ejemplo: $x + 5 = 10$ (Â¿quÃ© valor de $x$ satisface la ecuaciÃ³n?)

**En ComputaciÃ³n:** Una variable es un **nombre** que apunta a un objeto **conocido**  
- Ejemplo: `voltaje_entrada = 5.0` (asignamos el valor 5.0 al nombre voltaje_entrada)

### âš ï¸ El sÃ­mbolo `=` NO es igualdad, es **asignaciÃ³n**

### Reglas de Nomenclatura en Python:

**Obligatorias:**
- Empiezan con letra (`a-z`, `A-Z`) o guiÃ³n bajo (`_`)
- Solo pueden contener letras, nÃºmeros y guiones bajos
- Son sensibles a mayÃºsculas (`voltaje != Voltaje`)
- No pueden ser palabras reservadas (`if`, `for`, `def`, etc.)

**Convenciones (PEP 8):**
- Usar `snake_case`: minÃºsculas con guiones bajos
- Nombres descriptivos que expliquen el propÃ³sito
- Evitar abreviaciones confusas

## ğŸ’» Ejercicio PrÃ¡ctico 1: Calculadora de Propiedades de Cilindro

Vamos a crear un programa que calcule las propiedades de un cilindro usando lo que hemos aprendido:

**Objetivo:**
1. Declarar variables `radio` y `altura` (flotantes)
2. Calcular volumen: $V = \pi r^2 h$ 
3. Calcular Ã¡rea lateral: $A = 2 \pi r h$
4. Determinar si el volumen > 50 mÂ³ (booleano)

Â¡Modifica los valores y experimenta!

````python
# ğŸ§® CALCULADORA INTERACTIVA DE CILINDRO
# âœ¨ Cambia los valores de radio y altura para experimentar

import math

# ParÃ¡metros del cilindro - Â¡MODIFICA ESTOS VALORES!
radio = 3.5    # metros
altura = 8.0   # metros

print("ğŸ—ï¸ CALCULADORA DE PROPIEDADES DEL CILINDRO")
print("=" * 50)
print(f"ğŸ“ ParÃ¡metros de entrada:")
print(f"   Radio: {radio} m")
print(f"   Altura: {altura} m")

# CÃ¡lculos usando variables
pi = math.pi  # Valor mÃ¡s preciso de Ï€
volumen = pi * radio**2 * altura
area_lateral = 2 * pi * radio * altura
area_base = pi * radio**2
area_total = area_lateral + 2 * area_base

# EvaluaciÃ³n booleana
es_gran_capacidad = volumen > 50

print(f"\nğŸ“Š Resultados calculados:")
print(f"   Volumen: {volumen:.2f} mÂ³")
print(f"   Ãrea lateral: {area_lateral:.2f} mÂ²")
print(f"   Ãrea total: {area_total:.2f} mÂ²")
print(f"   Â¿Gran capacidad? {es_gran_capacidad}")

# VerificaciÃ³n de tipos
print(f"\nğŸ” VerificaciÃ³n de tipos:")
print(f"   radio: {type(radio).__name__}")
print(f"   volumen: {type(volumen).__name__}")
print(f"   es_gran_capacidad: {type(es_gran_capacidad).__name__}")

# InformaciÃ³n adicional
if es_gran_capacidad:
    print(f"\nâœ… Este cilindro tiene gran capacidad para almacenamiento")
else:
    print(f"\nğŸ“¦ Este cilindro es de capacidad estÃ¡ndar")
    
print(f"\nğŸ’¡ Â¡Cambia los valores de radio y altura arriba y ejecuta de nuevo!")
````

## ğŸ” VerificaciÃ³n de Tipos (type checking)

En programaciÃ³n, es fundamental saber quÃ© tipo de dato estamos manejando. Python nos proporciona herramientas para verificar y entender los tipos de nuestras variables.

### Â¿Por quÃ© es importante?
- **DepuraciÃ³n**: Identificar errores relacionados con tipos incorrectos
- **ValidaciÃ³n**: Asegurar que los datos sean del tipo esperado
- **DocumentaciÃ³n**: Hacer el cÃ³digo mÃ¡s claro y comprensible
- **Robustez**: Prevenir errores en tiempo de ejecuciÃ³n

### Herramientas de verificaciÃ³n:
- `type()`: Devuelve el tipo exacto
- `isinstance()`: Verifica si un objeto es de un tipo especÃ­fico
- `dir()`: Lista mÃ©todos y atributos disponibles

````python
# ğŸ”¬ EXPLORACIÃ“N DE TIPOS DE DATOS

# Creamos variables de diferentes tipos
nombre = "Ana"              # string (cadena de caracteres)
edad = 20                   # int (entero)
altura = 1.65               # float (flotante)
es_estudiante = True        # bool (booleano)

print("ğŸ§ª ANÃLISIS DE TIPOS DE DATOS")
print("=" * 40)

print(f"Nombre: {nombre}")
print(f"Tipo de nombre: {type(nombre)}")

print(f"\nEdad: {edad}")
print(f"Tipo de edad: {type(edad)}")

print(f"\nAltura: {altura}")
print(f"Tipo de altura: {type(altura)}")

print(f"\nEs estudiante: {es_estudiante}")
print(f"Tipo de es_estudiante: {type(es_estudiante)}")

# Verificaciones simples con type()
print(f"\nğŸ” VERIFICACIONES:")
print(f"Â¿La edad es un entero? {type(edad) == int}")
print(f"Â¿El nombre es una cadena? {type(nombre) == str}")
print(f"Â¿La altura es un flotante? {type(altura) == float}")
print(f"Â¿es_estudiante es booleano? {type(es_estudiante) == bool}")
````

## ğŸ”„ ConversiÃ³n de Tipos (Type Casting)

La conversiÃ³n de tipos nos permite transformar datos de un tipo a otro. Esto es especialmente Ãºtil cuando:
- Leemos datos de archivos (siempre llegan como strings)
- Procesamos entrada del usuario
- Necesitamos realizar operaciones entre tipos compatibles
- Preparamos datos para almacenamiento o transmisiÃ³n

### Funciones de conversiÃ³n principales:
- `int()`: Convierte a entero
- `float()`: Convierte a flotante  
- `str()`: Convierte a string
- `bool()`: Convierte a booleano

### âš ï¸ Importante:
Algunas conversiones pueden fallar si los datos no son compatibles.

````python
# ğŸ”„ LABORATORIO DE CONVERSIÃ“N DE TIPOS

print("ğŸ§ª LABORATORIO DE CONVERSIÃ“N DE TIPOS")
print("=" * 50)

# Datos de ejemplo
numero_como_texto = "25"        # String que representa un entero
decimal_como_texto = "3.14"     # String que representa un flotante
numero_entero = 42              # Entero que queremos como string
numero_decimal = 2.71           # Flotante que queremos como entero

print("ğŸ“Š DATOS ORIGINALES:")
print(f"numero_como_texto: '{numero_como_texto}' (tipo: {type(numero_como_texto).__name__})")
print(f"decimal_como_texto: '{decimal_como_texto}' (tipo: {type(decimal_como_texto).__name__})")
print(f"numero_entero: {numero_entero} (tipo: {type(numero_entero).__name__})")
print(f"numero_decimal: {numero_decimal} (tipo: {type(numero_decimal).__name__})")

print(f"\nğŸ”„ CONVERSIONES:")

# String a nÃºmero
numero = int(numero_como_texto)
print(f"String a int: '{numero_como_texto}' â†’ {numero}")

decimal = float(decimal_como_texto)
print(f"String a float: '{decimal_como_texto}' â†’ {decimal}")

# NÃºmero a string
texto = str(numero_entero)
print(f"Int a string: {numero_entero} â†’ '{texto}'")

# Float a int (pierde la parte decimal)
entero = int(numero_decimal)
print(f"Float a int: {numero_decimal} â†’ {entero}")

# Conversiones a booleano
print(f"\nğŸ” CONVERSIONES A BOOLEANO:")
print(f"bool(1): {bool(1)}")
print(f"bool(0): {bool(0)}")
print(f"bool('hola'): {bool('hola')}")
print(f"bool(''): {bool('')}")

# âš ï¸ Conversiones que pueden dar error
print(f"\nâš ï¸ CUIDADO CON ESTAS CONVERSIONES:")
try:
    numero_malo = int("hola")
except ValueError:
    print("int('hola') â†’ Â¡ERROR! No se puede convertir texto a nÃºmero")

try:
    decimal_malo = float("abc")
except ValueError:
    print("float('abc') â†’ Â¡ERROR! No se puede convertir texto a decimal")
````

## ğŸ“¥ Manejo de Entrada BÃ¡sica

En programaciÃ³n frecuentemente necesitamos:
- Trabajar con datos que llegan como texto
- Convertir texto a nÃºmeros para hacer cÃ¡lculos
- Validar que los datos sean correctos

### Conceptos importantes:
- Los datos que leemos siempre llegan como strings (texto)

### ğŸ’¡ Nota sobre input():
En notebooks, simularemos entrada de usuario con datos predefinidos.

````python
# ğŸ“Š EJERCICIO: TRABAJANDO CON DATOS DE TEXTO
# Simulamos datos que llegan como texto y necesitamos convertir

print("ğŸ”¬ TRABAJANDO CON DATOS DE TEXTO")
print("=" * 50)

# Simulamos datos que llegan como strings (como del usuario o un archivo)
edad_texto = "25"
altura_texto = "1.75"
nombre = "Carlos"
es_estudiante_texto = "true"

print("ğŸ“„ DATOS COMO TEXTO:")
print(f"Edad: '{edad_texto}' (tipo: {type(edad_texto).__name__})")
print(f"Altura: '{altura_texto}' (tipo: {type(altura_texto).__name__})")
print(f"Nombre: '{nombre}' (tipo: {type(nombre).__name__})")
print(f"Es estudiante: '{es_estudiante_texto}' (tipo: {type(es_estudiante_texto).__name__})")

# Convertir a los tipos correctos
print(f"\nğŸ”„ CONVIRTIENDO A TIPOS CORRECTOS:")

edad = int(edad_texto)
print(f"Edad como nÃºmero: {edad} (tipo: {type(edad).__name__})")

altura = float(altura_texto)
print(f"Altura como decimal: {altura} (tipo: {type(altura).__name__})")

# Para booleanos, necesitamos decidir quÃ© significa "true"
es_estudiante = es_estudiante_texto.lower() == "true"
print(f"Es estudiante: {es_estudiante} (tipo: {type(es_estudiante).__name__})")

# Ahora podemos hacer cÃ¡lculos
print(f"\nğŸ“Š CÃLCULOS CON LOS DATOS CONVERTIDOS:")
print(f"{nombre} tiene {edad} aÃ±os")
print(f"Su altura es {altura} metros")

# CÃ¡lculo del IMC (solo como ejemplo de usar los nÃºmeros)
peso = 70  # kg (valor de ejemplo)
imc = peso / (altura ** 2)
print(f"Su IMC aproximado es: {imc:.1f}")

# Verificar si es mayor de edad
es_mayor_edad = edad >= 18
print(f"Â¿Es mayor de edad? {es_mayor_edad}")

print(f"\nğŸ’¡ Los datos originales eran texto, pero ahora podemos usarlos para cÃ¡lculos!")
````

## ğŸ”’ Mutabilidad y Identidad de Objetos

Concepto fundamental en Python: algunos tipos de datos pueden modificarse despuÃ©s de su creaciÃ³n (mutables) mientras que otros no (inmutables).

### Tipos inmutables:
- `int`, `float`, `bool`: Los nÃºmeros no cambian
- `str`: Las cadenas no se modifican, se crean nuevas
- `tuple`: Tuplas son inmutables

### Tipos mutables:
- `list`: Las listas pueden modificarse
- `dict`: Los diccionarios pueden cambiar
- `set`: Los conjuntos son modificables

### Â¿Por quÃ© importa?
- **Rendimiento**: Evitar copias innecesarias
- **Comportamiento**: Entender cuÃ¡ndo las variables "comparten" datos
- **Debugging**: Cambios inesperados en los datos
- **DiseÃ±o**: Elegir el tipo correcto para cada situaciÃ³n

````python
# ğŸ”¬ DEMOSTRACIÃ“N BÃSICA DE MUTABILIDAD

print("ğŸ”¬ CONCEPTO DE MUTABILIDAD")
print("=" * 40)

# Ejemplo con tipos INMUTABLES (nÃºmeros y strings)
print("ğŸ“Š TIPOS INMUTABLES:")
print("   Los nÃºmeros y strings NO se pueden cambiar")

numero = 42
print(f"NÃºmero original: {numero}")

# Si "modificamos" un nÃºmero, realmente creamos uno nuevo
numero = numero + 1
print(f"NÃºmero despuÃ©s de +1: {numero}")
print("ğŸ’¡ Se creÃ³ un nuevo nÃºmero, el original no cambiÃ³")

# Strings tambiÃ©n son inmutables
texto = "Hola"
print(f"\nTexto original: '{texto}'")

# Si "modificamos" un string, creamos uno nuevo
texto = texto + " mundo"
print(f"Texto despuÃ©s de concatenar: '{texto}'")
print("ğŸ’¡ Se creÃ³ un nuevo string, el original no cambiÃ³")

# Ejemplo simple con listas (MUTABLES)
print(f"\nğŸ“Š TIPOS MUTABLES:")
print("   Las listas SÃ se pueden cambiar")

lista_numeros = [1, 2, 3]
print(f"Lista original: {lista_numeros}")

# Modificamos la lista original
lista_numeros.append(4)
print(f"Lista despuÃ©s de agregar 4: {lista_numeros}")
print("ğŸ’¡ La misma lista cambiÃ³, no se creÃ³ una nueva")

# DemostraciÃ³n con variables que apuntan a la misma lista
print(f"\nâš ï¸ CUIDADO CON LAS LISTAS:")

lista_a = [1, 2, 3]
lista_b = lista_a  # Â¡Ambas variables apuntan a la MISMA lista!

print(f"lista_a: {lista_a}")
print(f"lista_b: {lista_b}")

# Modificamos a travÃ©s de lista_b
lista_b.append(4)
print(f"\nDespuÃ©s de lista_b.append(4):")
print(f"lista_a: {lista_a}")  # Â¡TambiÃ©n cambiÃ³!
print(f"lista_b: {lista_b}")
print("ğŸ’¡ Ambas variables apuntan a la misma lista")

print(f"\nğŸ¯ RESUMEN:")
print("â€¢ int, float, bool, str â†’ INMUTABLES (no cambian)")
print("â€¢ list, dict â†’ MUTABLES (sÃ­ cambian)")
print("â€¢ Â¡Cuidado al compartir listas entre variables!")
````

````python
# ğŸ¯ EJERCICIO FINAL: CALCULADORA PERSONAL
# Â¡Practica todos los conceptos bÃ¡sicos que has aprendido!

print("ğŸ¯ EJERCICIO FINAL: CALCULADORA PERSONAL")
print("=" * 50)

# Simulamos datos de una persona (como si vinieran del usuario)
nombre_completo = "MarÃ­a GarcÃ­a"
edad_texto = "22"
altura_texto = "1.68"
peso_texto = "58.5"
es_estudiante_texto = "sÃ­"

print("ğŸ“„ DATOS PERSONALES (como texto):")
print(f"Nombre: {nombre_completo}")
print(f"Edad: {edad_texto}")
print(f"Altura: {altura_texto}")
print(f"Peso: {peso_texto}")
print(f"Â¿Es estudiante?: {es_estudiante_texto}")

# Verificar tipos originales
print(f"\nğŸ” TIPOS ORIGINALES:")
print(f"Tipo de edad_texto: {type(edad_texto)}")
print(f"Tipo de altura_texto: {type(altura_texto)}")
print(f"Tipo de peso_texto: {type(peso_texto)}")

# Convertir a tipos apropiados
print(f"\nğŸ”„ CONVIRTIENDO A TIPOS CORRECTOS:")

edad = int(edad_texto)
altura = float(altura_texto)
peso = float(peso_texto)
es_estudiante = es_estudiante_texto.lower() == "sÃ­"

print(f"Edad convertida: {edad} (tipo: {type(edad).__name__})")
print(f"Altura convertida: {altura} (tipo: {type(altura).__name__})")
print(f"Peso convertido: {peso} (tipo: {type(peso).__name__})")
print(f"Es estudiante: {es_estudiante} (tipo: {type(es_estudiante).__name__})")

# Hacer algunos cÃ¡lculos bÃ¡sicos
print(f"\nğŸ“Š CÃLCULOS PERSONALES:")

# Calcular IMC
imc = peso / (altura * altura)
print(f"Ãndice de Masa Corporal (IMC): {imc:.2f}")

# Determinar estado del IMC (usando booleanos)
peso_bajo = imc < 18.5
peso_normal = 18.5 <= imc < 25
sobrepeso = 25 <= imc < 30
obesidad = imc >= 30

print(f"Â¿Peso bajo? {peso_bajo}")
print(f"Â¿Peso normal? {peso_normal}")
print(f"Â¿Sobrepeso? {sobrepeso}")
print(f"Â¿Obesidad? {obesidad}")

# InformaciÃ³n adicional
print(f"\nğŸ‘¤ INFORMACIÃ“N PERSONAL:")
print(f"{nombre_completo} tiene {edad} aÃ±os")
print(f"Altura: {altura} metros")
print(f"Peso: {peso} kg")

# Verificar mayorÃ­a de edad
mayor_edad = edad >= 18
print(f"Â¿Es mayor de edad? {mayor_edad}")

# Estado acadÃ©mico
if es_estudiante:
    print("Estado: Estudiante activo")
else:
    print("Estado: No estudiante")

# Mostrar categorÃ­a de IMC
print(f"\nCategorÃ­a de IMC:")
if peso_bajo:
    categoria = "Peso bajo"
elif peso_normal:
    categoria = "Peso normal"
elif sobrepeso:
    categoria = "Sobrepeso"
else:
    categoria = "Obesidad"

print(f"CategorÃ­a: {categoria}")

print(f"\nğŸ‰ Â¡EJERCICIO COMPLETADO!")
print("Has aplicado exitosamente:")
print("âœ… Trabajo con strings (cadenas de caracteres)")
print("âœ… ConversiÃ³n de tipos (str â†’ int, str â†’ float)")
print("âœ… Uso de booleanos para condiciones")
print("âœ… Operaciones matemÃ¡ticas con flotantes")
print("âœ… AsignaciÃ³n y uso de variables")
print("âœ… VerificaciÃ³n de tipos de datos")

print(f"\nğŸ’¡ Â¡Modifica los datos arriba para probar con diferentes personas!")
````
