## üî¢ Secci√≥n 2: Trabajando con Enteros (`int`)

- Pueden ser positivos, negativos o cero
- Perfectos para conteos, √≠ndices, y c√°lculos exactos
- En Python 3, pueden ser arbitrariamente grandes

### Ejemplos de uso en ingenier√≠a:

## üåä Secci√≥n 3: Trabajando con Flotantes (`float`)

Los **flotantes** representan n√∫meros reales con **precisi√≥n finita** - pueden tener parte decimal.

### Caracter√≠sticas principales:
- Tienen parte decimal (pueden ser 0.0)
- Representaci√≥n de punto flotante IEEE 754
- Precisi√≥n limitada (~15-17 d√≠gitos significativos)
- Pueden usar notaci√≥n cient√≠fica (ej: `1.5e-10`)

### Ejemplos de uso en ingenier√≠a:
- Medidas de sensores (voltaje, temperatura, presi√≥n)
- Constantes f√≠sicas (œÄ, e, constante de Planck)
- C√°lculos de precisi√≥n (coordenadas GPS)
- Algoritmos num√©ricos

````python
# Experimentando con flotantes
print("üåä TRABAJANDO CON FLOTANTES")
print("=" * 30)

# Ejemplos b√°sicos
voltaje = 4.95
temperatura = 23.7
pi_aproximado = 3.14159
constante_planck = 6.626e-34  # Notaci√≥n cient√≠fica

print(f"Voltaje del sensor: {voltaje} V")
print(f"Temperatura ambiente: {temperatura} ¬∞C")  
print(f"Pi aproximado: {pi_aproximado}")
print(f"Constante de Planck: {constante_planck} J‚ãÖs")

# Verificar tipos
print(f"\nTipo de voltaje: {type(voltaje)}")
print(f"Tipo de constante_planck: {type(constante_planck)}")

# Operaciones matem√°ticas
print(f"\nOperaciones con flotantes:")
print(f"Temperatura en Fahrenheit: {temperatura * 9/5 + 32:.1f} ¬∞F")
print(f"√Årea de c√≠rculo (r=2.5): {pi_aproximado * 2.5**2:.2f} m¬≤")
print(f"Voltaje RMS: {voltaje / 1.414:.3f} V")

# Demostraci√≥n de precisi√≥n
print(f"\nPrecisi√≥n de flotantes:")
print(f"0.1 + 0.2 = {0.1 + 0.2}")  # ¬°Sorpresa!
print(f"¬øEs 0.1 + 0.2 == 0.3? {0.1 + 0.2 == 0.3}")
print("üí° Esto demuestra la limitaci√≥n de precisi√≥n de los flotantes")
````

## ‚úÖ‚ùå Secci√≥n 4: Trabajando con Booleanos (`bool`)

Los **booleanos** son el tipo m√°s simple: solo pueden ser `True` o `False`.

### Caracter√≠sticas principales:
- Solo dos valores posibles: `True` y `False`
- Internamente, `True` es `1` y `False` es `0`
- Resultado de operaciones de comparaci√≥n
- Fundamentales para la l√≥gica de control

### Ejemplos de uso en ingenier√≠a:
- Estados de sensores (activo/inactivo)
- Condiciones de alarma (normal/cr√≠tico)
- Flags de sistema (encendido/apagado)
- Validaci√≥n de datos (v√°lido/inv√°lido)

````python
# Experimentando con booleanos
print("‚úÖ‚ùå TRABAJANDO CON BOOLEANOS")
print("=" * 30)

# Valores b√°sicos
verdadero = True
falso = False

print(f"Valor verdadero: {verdadero}")
print(f"Valor falso: {falso}")
print(f"Tipo de verdadero: {type(verdadero)}")

# Booleanos como n√∫meros
print(f"\nBooleanos como n√∫meros:")
print(f"True como n√∫mero: {True + 0}")
print(f"False como n√∫mero: {False + 0}")
print(f"Suma True + True: {True + True}")

# Comparaciones que generan booleanos
temperatura_actual = 95.5
temperatura_maxima = 90.0

print(f"\nüå°Ô∏è Sistema de Monitoreo de Temperatura:")
print(f"Temperatura actual: {temperatura_actual}¬∞C")
print(f"Temperatura m√°xima: {temperatura_maxima}¬∞C")

# Operaciones de comparaci√≥n
alarma_activada = temperatura_actual > temperatura_maxima
temperatura_normal = temperatura_actual <= temperatura_maxima
temperatura_critica = temperatura_actual > 100.0

print(f"\nEstados del sistema:")
print(f"¬øAlarma activada? {alarma_activada}")
print(f"¬øTemperatura normal? {temperatura_normal}")
print(f"¬øTemperatura cr√≠tica? {temperatura_critica}")

# Operadores l√≥gicos
sistema_seguro = temperatura_normal and not temperatura_critica
print(f"¬øSistema seguro? {sistema_seguro}")
````

- Se crean con comillas simples `'...'`, dobles `"..."` o triples `'''...'''`
- Soportan operaciones como concatenaci√≥n, repetici√≥n y slicing

### Ejemplos de uso en ingenier√≠a:

- C√≥digos de identificaci√≥n

- Comandos de comunicaci√≥n

````python
# Experimentando con cadenas
print("üìù TRABAJANDO CON CADENAS")
print("=" * 30)

# Creaci√≥n de cadenas
nombre_sensor = "Sensor de Temperatura DHT22"
id_componente = 'X-48-AB-v2'
letra = 'A'  # Un car√°cter es una cadena de longitud 1
cadena_vacia = ""

print(f"Nombre del sensor: {nombre_sensor}")
print(f"ID del componente: {id_componente}")
print(f"Una letra: {letra}")
print(f"Cadena vac√≠a: '{cadena_vacia}'")

# Verificar tipos y longitudes
print(f"\nPropiedades:")
print(f"Tipo de nombre_sensor: {type(nombre_sensor)}")
print(f"Longitud del nombre: {len(nombre_sensor)} caracteres")
print(f"Longitud de una letra: {len(letra)} caracteres")

# Operaciones con cadenas
print(f"\nOperaciones con cadenas:")
print(f"Concatenaci√≥n: {id_componente + '-' + letra}")
print(f"Repetici√≥n: {letra * 5}")
print(f"May√∫sculas: {nombre_sensor.upper()}")
print(f"Min√∫sculas: {id_componente.lower()}")

# Indexaci√≥n y slicing
print(f"\nAcceso a caracteres:")
print(f"Primer car√°cter: '{nombre_sensor[0]}'")
print(f"√öltimo car√°cter: '{nombre_sensor[-1]}'")
print(f"Primeras 6 letras: '{nombre_sensor[:6]}'")
print(f"Palabra 'Sensor': '{nombre_sensor[0:6]}'")

# M√©todos √∫tiles para ingenier√≠a
codigo_sensor = "TEMP_01_ACTIVO"
print(f"\nüîß Procesamiento de c√≥digos:")
print(f"C√≥digo original: {codigo_sensor}")
print(f"¬øContiene 'TEMP'? {codigo_sensor.startswith('TEMP')}")
print(f"¬øEst√° activo? {codigo_sensor.endswith('ACTIVO')}")
print(f"Separar por '_': {codigo_sensor.split('_')}")
````

## üè∑Ô∏è Secci√≥n 6: Asignaci√≥n de Variables y Nomenclatura

### Diferencia clave: Matem√°ticas vs. Computaci√≥n

**En Matem√°ticas:** Una variable es un s√≠mbolo para un valor **desconocido**  
- Ejemplo: $x + 5 = 10$ (¬øqu√© valor de $x$ satisface la ecuaci√≥n?)

**En Computaci√≥n:** Una variable es un **nombre** que apunta a un objeto **conocido**  
- Ejemplo: `voltaje_entrada = 5.0` (asignamos el valor 5.0 al nombre voltaje_entrada)

### ‚ö†Ô∏è El s√≠mbolo `=` NO es igualdad, es **asignaci√≥n**

### Reglas de Nomenclatura en Python:

**Obligatorias:**
- Empiezan con letra (`a-z`, `A-Z`) o gui√≥n bajo (`_`)
- Solo pueden contener letras, n√∫meros y guiones bajos
- Son sensibles a may√∫sculas (`voltaje != Voltaje`)
- No pueden ser palabras reservadas (`if`, `for`, `def`, etc.)

**Convenciones (PEP 8):**
- Usar `snake_case`: min√∫sculas con guiones bajos
- Nombres descriptivos que expliquen el prop√≥sito
- Evitar abreviaciones confusas

## üíª Ejercicio Pr√°ctico 1: Calculadora de Propiedades de Cilindro

Vamos a crear un programa que calcule las propiedades de un cilindro usando lo que hemos aprendido:

**Objetivo:**
1. Declarar variables `radio` y `altura` (flotantes)
2. Calcular volumen: $V = \pi r^2 h$ 
3. Calcular √°rea lateral: $A = 2 \pi r h$
4. Determinar si el volumen > 50 m¬≥ (booleano)

¬°Modifica los valores y experimenta!

````python
# üßÆ CALCULADORA INTERACTIVA DE CILINDRO
# ‚ú® Cambia los valores de radio y altura para experimentar

import math

# Par√°metros del cilindro - ¬°MODIFICA ESTOS VALORES!
radio = 3.5    # metros
altura = 8.0   # metros

print("üèóÔ∏è CALCULADORA DE PROPIEDADES DEL CILINDRO")
print("=" * 50)
print(f"üìê Par√°metros de entrada:")
print(f"   Radio: {radio} m")
print(f"   Altura: {altura} m")

# C√°lculos usando variables
pi = math.pi  # Valor m√°s preciso de œÄ
volumen = pi * radio**2 * altura
area_lateral = 2 * pi * radio * altura
area_base = pi * radio**2
area_total = area_lateral + 2 * area_base

# Evaluaci√≥n booleana
es_gran_capacidad = volumen > 50

print(f"\nüìä Resultados calculados:")
print(f"   Volumen: {volumen:.2f} m¬≥")
print(f"   √Årea lateral: {area_lateral:.2f} m¬≤")
print(f"   √Årea total: {area_total:.2f} m¬≤")
print(f"   ¬øGran capacidad? {es_gran_capacidad}")

# Verificaci√≥n de tipos
print(f"\nüîç Verificaci√≥n de tipos:")
print(f"   radio: {type(radio).__name__}")
print(f"   volumen: {type(volumen).__name__}")
print(f"   es_gran_capacidad: {type(es_gran_capacidad).__name__}")

# Informaci√≥n adicional
if es_gran_capacidad:
    print(f"\n‚úÖ Este cilindro tiene gran capacidad para almacenamiento")
else:
    print(f"\nüì¶ Este cilindro es de capacidad est√°ndar")
    
print(f"\nüí° ¬°Cambia los valores de radio y altura arriba y ejecuta de nuevo!")
````

## üîç Verificaci√≥n de Tipos (type checking)

En programaci√≥n, es fundamental saber qu√© tipo de dato estamos manejando. Python nos proporciona herramientas para verificar y entender los tipos de nuestras variables.

### ¬øPor qu√© es importante?
- **Depuraci√≥n**: Identificar errores relacionados con tipos incorrectos
- **Validaci√≥n**: Asegurar que los datos sean del tipo esperado
- **Documentaci√≥n**: Hacer el c√≥digo m√°s claro y comprensible
- **Robustez**: Prevenir errores en tiempo de ejecuci√≥n

### Herramientas de verificaci√≥n:
- `type()`: Devuelve el tipo exacto
- `isinstance()`: Verifica si un objeto es de un tipo espec√≠fico
- `dir()`: Lista m√©todos y atributos disponibles

````python
# üî¨ EXPLORACI√ìN DE TIPOS DE DATOS

# Creamos variables de diferentes tipos
nombre = "Ana"              # string (cadena de caracteres)
edad = 20                   # int (entero)
altura = 1.65               # float (flotante)
es_estudiante = True        # bool (booleano)

print("üß™ AN√ÅLISIS DE TIPOS DE DATOS")
print("=" * 40)

print(f"Nombre: {nombre}")
print(f"Tipo de nombre: {type(nombre)}")

print(f"\nEdad: {edad}")
print(f"Tipo de edad: {type(edad)}")

print(f"\nAltura: {altura}")
print(f"Tipo de altura: {type(altura)}")

print(f"\nEs estudiante: {es_estudiante}")
print(f"Tipo de es_estudiante: {type(es_estudiante)}")

# Verificaciones simples con type()
print(f"\nüîç VERIFICACIONES:")
print(f"¬øLa edad es un entero? {type(edad) == int}")
print(f"¬øEl nombre es una cadena? {type(nombre) == str}")
print(f"¬øLa altura es un flotante? {type(altura) == float}")
print(f"¬øes_estudiante es booleano? {type(es_estudiante) == bool}")
````

## üîÑ Conversi√≥n de Tipos (Type Casting)

La conversi√≥n de tipos nos permite transformar datos de un tipo a otro. Esto es especialmente √∫til cuando:
- Leemos datos de archivos (siempre llegan como strings)
- Procesamos entrada del usuario
- Necesitamos realizar operaciones entre tipos compatibles
- Preparamos datos para almacenamiento o transmisi√≥n

### Funciones de conversi√≥n principales:
- `int()`: Convierte a entero
- `float()`: Convierte a flotante  
- `str()`: Convierte a string
- `bool()`: Convierte a booleano

### ‚ö†Ô∏è Importante:
Algunas conversiones pueden fallar si los datos no son compatibles.

````python
# üîÑ LABORATORIO DE CONVERSI√ìN DE TIPOS

print("üß™ LABORATORIO DE CONVERSI√ìN DE TIPOS")
print("=" * 50)

# Datos de ejemplo
numero_como_texto = "25"        # String que representa un entero
decimal_como_texto = "3.14"     # String que representa un flotante
numero_entero = 42              # Entero que queremos como string
numero_decimal = 2.71           # Flotante que queremos como entero

print("üìä DATOS ORIGINALES:")
print(f"numero_como_texto: '{numero_como_texto}' (tipo: {type(numero_como_texto).__name__})")
print(f"decimal_como_texto: '{decimal_como_texto}' (tipo: {type(decimal_como_texto).__name__})")
print(f"numero_entero: {numero_entero} (tipo: {type(numero_entero).__name__})")
print(f"numero_decimal: {numero_decimal} (tipo: {type(numero_decimal).__name__})")

print(f"\nüîÑ CONVERSIONES:")

# String a n√∫mero
numero = int(numero_como_texto)
print(f"String a int: '{numero_como_texto}' ‚Üí {numero}")

decimal = float(decimal_como_texto)
print(f"String a float: '{decimal_como_texto}' ‚Üí {decimal}")

# N√∫mero a string
texto = str(numero_entero)
print(f"Int a string: {numero_entero} ‚Üí '{texto}'")

# Float a int (pierde la parte decimal)
entero = int(numero_decimal)
print(f"Float a int: {numero_decimal} ‚Üí {entero}")

# Conversiones a booleano
print(f"\nüîç CONVERSIONES A BOOLEANO:")
print(f"bool(1): {bool(1)}")
print(f"bool(0): {bool(0)}")
print(f"bool('hola'): {bool('hola')}")
print(f"bool(''): {bool('')}")

# ‚ö†Ô∏è Conversiones que pueden dar error
print(f"\n‚ö†Ô∏è CUIDADO CON ESTAS CONVERSIONES:")
try:
    numero_malo = int("hola")
except ValueError:
    print("int('hola') ‚Üí ¬°ERROR! No se puede convertir texto a n√∫mero")

try:
    decimal_malo = float("abc")
except ValueError:
    print("float('abc') ‚Üí ¬°ERROR! No se puede convertir texto a decimal")
````

## üì• Manejo de Entrada B√°sica

En programaci√≥n frecuentemente necesitamos:
- Trabajar con datos que llegan como texto
- Convertir texto a n√∫meros para hacer c√°lculos
- Validar que los datos sean correctos

### Conceptos importantes:
- Los datos que leemos siempre llegan como strings (texto)
- Necesitamos convertirlos al tipo correcto para trabajar con ellos
- Debemos manejar casos donde la conversi√≥n no es posible

### üí° Nota sobre input():
En notebooks, simularemos entrada de usuario con datos predefinidos.

````python
# üìä EJERCICIO: TRABAJANDO CON DATOS DE TEXTO
# Simulamos datos que llegan como texto y necesitamos convertir

print("üî¨ TRABAJANDO CON DATOS DE TEXTO")
print("=" * 50)

# Simulamos datos que llegan como strings (como del usuario o un archivo)
edad_texto = "25"
altura_texto = "1.75"
nombre = "Carlos"
es_estudiante_texto = "true"

print("üìÑ DATOS COMO TEXTO:")
print(f"Edad: '{edad_texto}' (tipo: {type(edad_texto).__name__})")
print(f"Altura: '{altura_texto}' (tipo: {type(altura_texto).__name__})")
print(f"Nombre: '{nombre}' (tipo: {type(nombre).__name__})")
print(f"Es estudiante: '{es_estudiante_texto}' (tipo: {type(es_estudiante_texto).__name__})")

# Convertir a los tipos correctos
print(f"\nüîÑ CONVIRTIENDO A TIPOS CORRECTOS:")

edad = int(edad_texto)
print(f"Edad como n√∫mero: {edad} (tipo: {type(edad).__name__})")

altura = float(altura_texto)
print(f"Altura como decimal: {altura} (tipo: {type(altura).__name__})")

# Para booleanos, necesitamos decidir qu√© significa "true"
es_estudiante = es_estudiante_texto.lower() == "true"
print(f"Es estudiante: {es_estudiante} (tipo: {type(es_estudiante).__name__})")

# Ahora podemos hacer c√°lculos
print(f"\nüìä C√ÅLCULOS CON LOS DATOS CONVERTIDOS:")
print(f"{nombre} tiene {edad} a√±os")
print(f"Su altura es {altura} metros")

# C√°lculo del IMC (solo como ejemplo de usar los n√∫meros)
peso = 70  # kg (valor de ejemplo)
imc = peso / (altura ** 2)
print(f"Su IMC aproximado es: {imc:.1f}")

# Verificar si es mayor de edad
es_mayor_edad = edad >= 18
print(f"¬øEs mayor de edad? {es_mayor_edad}")

print(f"\nüí° Los datos originales eran texto, pero ahora podemos usarlos para c√°lculos!")
````

## üîí Mutabilidad y Identidad de Objetos

Concepto fundamental en Python: algunos tipos de datos pueden modificarse despu√©s de su creaci√≥n (mutables) mientras que otros no (inmutables).

### Tipos inmutables:
- `int`, `float`, `bool`: Los n√∫meros no cambian
- `str`: Las cadenas no se modifican, se crean nuevas
- `tuple`: Tuplas son inmutables

### Tipos mutables:
- `list`: Las listas pueden modificarse
- `dict`: Los diccionarios pueden cambiar
- `set`: Los conjuntos son modificables

### ¬øPor qu√© importa?
- **Rendimiento**: Evitar copias innecesarias
- **Comportamiento**: Entender cu√°ndo las variables "comparten" datos
- **Debugging**: Cambios inesperados en los datos
- **Dise√±o**: Elegir el tipo correcto para cada situaci√≥n

````python
# üî¨ DEMOSTRACI√ìN B√ÅSICA DE MUTABILIDAD

print("üî¨ CONCEPTO DE MUTABILIDAD")
print("=" * 40)

# Ejemplo con tipos INMUTABLES (n√∫meros y strings)
print("üìä TIPOS INMUTABLES:")
print("   Los n√∫meros y strings NO se pueden cambiar")

numero = 42
print(f"N√∫mero original: {numero}")

# Si "modificamos" un n√∫mero, realmente creamos uno nuevo
numero = numero + 1
print(f"N√∫mero despu√©s de +1: {numero}")
print("üí° Se cre√≥ un nuevo n√∫mero, el original no cambi√≥")

# Strings tambi√©n son inmutables
texto = "Hola"
print(f"\nTexto original: '{texto}'")

# Si "modificamos" un string, creamos uno nuevo
texto = texto + " mundo"
print(f"Texto despu√©s de concatenar: '{texto}'")
print("üí° Se cre√≥ un nuevo string, el original no cambi√≥")

# Ejemplo simple con listas (MUTABLES)
print(f"\nüìä TIPOS MUTABLES:")
print("   Las listas S√ç se pueden cambiar")

lista_numeros = [1, 2, 3]
print(f"Lista original: {lista_numeros}")

# Modificamos la lista original
lista_numeros.append(4)
print(f"Lista despu√©s de agregar 4: {lista_numeros}")
print("üí° La misma lista cambi√≥, no se cre√≥ una nueva")

# Demostraci√≥n con variables que apuntan a la misma lista
print(f"\n‚ö†Ô∏è CUIDADO CON LAS LISTAS:")

lista_a = [1, 2, 3]
lista_b = lista_a  # ¬°Ambas variables apuntan a la MISMA lista!

print(f"lista_a: {lista_a}")
print(f"lista_b: {lista_b}")

# Modificamos a trav√©s de lista_b
lista_b.append(4)
print(f"\nDespu√©s de lista_b.append(4):")
print(f"lista_a: {lista_a}")  # ¬°Tambi√©n cambi√≥!
print(f"lista_b: {lista_b}")
print("üí° Ambas variables apuntan a la misma lista")

print(f"\nüéØ RESUMEN:")
print("‚Ä¢ int, float, bool, str ‚Üí INMUTABLES (no cambian)")
print("‚Ä¢ list, dict ‚Üí MUTABLES (s√≠ cambian)")
print("‚Ä¢ ¬°Cuidado al compartir listas entre variables!")
````

````python
# üéØ EJERCICIO FINAL: CALCULADORA PERSONAL
# ¬°Practica todos los conceptos b√°sicos que has aprendido!

print("üéØ EJERCICIO FINAL: CALCULADORA PERSONAL")
print("=" * 50)

# Simulamos datos de una persona (como si vinieran del usuario)
nombre_completo = "Mar√≠a Garc√≠a"
edad_texto = "22"
altura_texto = "1.68"
peso_texto = "58.5"
es_estudiante_texto = "s√≠"

print("üìÑ DATOS PERSONALES (como texto):")
print(f"Nombre: {nombre_completo}")
print(f"Edad: {edad_texto}")
print(f"Altura: {altura_texto}")
print(f"Peso: {peso_texto}")
print(f"¬øEs estudiante?: {es_estudiante_texto}")

# Verificar tipos originales
print(f"\nüîç TIPOS ORIGINALES:")
print(f"Tipo de edad_texto: {type(edad_texto)}")
print(f"Tipo de altura_texto: {type(altura_texto)}")
print(f"Tipo de peso_texto: {type(peso_texto)}")

# Convertir a tipos apropiados
print(f"\nüîÑ CONVIRTIENDO A TIPOS CORRECTOS:")

edad = int(edad_texto)
altura = float(altura_texto)
peso = float(peso_texto)
es_estudiante = es_estudiante_texto.lower() == "s√≠"

print(f"Edad convertida: {edad} (tipo: {type(edad).__name__})")
print(f"Altura convertida: {altura} (tipo: {type(altura).__name__})")
print(f"Peso convertido: {peso} (tipo: {type(peso).__name__})")
print(f"Es estudiante: {es_estudiante} (tipo: {type(es_estudiante).__name__})")

# Hacer algunos c√°lculos b√°sicos
print(f"\nüìä C√ÅLCULOS PERSONALES:")

# Calcular IMC
imc = peso / (altura * altura)
print(f"√çndice de Masa Corporal (IMC): {imc:.2f}")

# Determinar estado del IMC (usando booleanos)
peso_bajo = imc < 18.5
peso_normal = 18.5 <= imc < 25
sobrepeso = 25 <= imc < 30
obesidad = imc >= 30

print(f"¬øPeso bajo? {peso_bajo}")
print(f"¬øPeso normal? {peso_normal}")
print(f"¬øSobrepeso? {sobrepeso}")
print(f"¬øObesidad? {obesidad}")

# Informaci√≥n adicional
print(f"\nüë§ INFORMACI√ìN PERSONAL:")
print(f"{nombre_completo} tiene {edad} a√±os")
print(f"Altura: {altura} metros")
print(f"Peso: {peso} kg")

# Verificar mayor√≠a de edad
mayor_edad = edad >= 18
print(f"¬øEs mayor de edad? {mayor_edad}")

# Estado acad√©mico
if es_estudiante:
    print("Estado: Estudiante activo")
else:
    print("Estado: No estudiante")

# Mostrar categor√≠a de IMC
print(f"\nCategor√≠a de IMC:")
if peso_bajo:
    categoria = "Peso bajo"
elif peso_normal:
    categoria = "Peso normal"
elif sobrepeso:
    categoria = "Sobrepeso"
else:
    categoria = "Obesidad"

print(f"Categor√≠a: {categoria}")

print(f"\nüéâ ¬°EJERCICIO COMPLETADO!")
print("Has aplicado exitosamente:")
print("‚úÖ Trabajo con strings (cadenas de caracteres)")
print("‚úÖ Conversi√≥n de tipos (str ‚Üí int, str ‚Üí float)")
print("‚úÖ Uso de booleanos para condiciones")
print("‚úÖ Operaciones matem√°ticas con flotantes")
print("‚úÖ Asignaci√≥n y uso de variables")
print("‚úÖ Verificaci√≥n de tipos de datos")

print(f"\nüí° ¬°Modifica los datos arriba para probar con diferentes personas!")
````
