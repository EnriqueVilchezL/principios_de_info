````python
sopa = [
    ["P","Y","T","H","O","N","D","R","Y","H"],
    ["C","U","O","G","I","D","O","C","I","Y"],
    ["S","U","V","Q","M","O","D","L","W","R"],
    ["O","D","A","E","S","U","E","P","Z","V"],
    ["J","L","L","I","N","R","U","N","R","Q"],
    ["A","K","O","X","A","T","G","D","R","S"],
    ["B","O","R","O","O","L","E","U","B","V"],
    ["A","I","G","T","B","F","O","R","N","A"],
    ["R","Y","A","S","A","V","G","O","O","C"],
    ["T","D","X","G","Q","S","C","S","B","B"]
]
````

````python
def encontrar_palabra_horizontal(sopa: list, palabra: str) -> tuple:
    """
    Busca una palabra en la sopa de letras de forma horizontal.

    Args:
        sopa (list): Matriz que representa la sopa de letras.
        palabra (str): Palabra a buscar en la sopa de letras.

    Returns:
        tuple: Una tupla que contiene la posici√≥n de inicio (fila, columna),
               la posici√≥n final (fila, columna) y la direcci√≥n ('‚Üí' o '‚Üê').
               Si no se encuentra la palabra, devuelve (None, None, None).
    """
    # Variable que indica la posici√≥n de inicio de la palabra
    posicion_inicio = None
    # Variable que indica la posici√≥n final de la palabra
    posicion_final = None
    # Variable que indica la direcci√≥n de la palabra
    direccion = None
    # Variable que indica si la palabra fue encontrada
    encontrada = False

    for fila in range(len(sopa)):
        # Si ya se encontr√≥ la palabra, no seguir buscando
        if not encontrada:

            # Convertir la fila de caracteres a una cadena para facilitar la b√∫squeda
            hilera_fila = ""
            for col in range(len(sopa[fila])):
                hilera_fila += sopa[fila][col]
            
            # Buscar la palabra en la hilera (de izquierda a derecha)
            indice = hilera_fila.find(palabra)

            # Si el √≠ndice es distinto de -1, la palabra fue encontrada
            if indice != -1:
                posicion_inicio = (fila, indice)
                posicion_final = (fila, indice + len(palabra) - 1)
                direccion = '‚Üí'
                encontrada = True

            else:
                # Si no se encuentra, buscar la palabra al rev√©s (de derecha a izquierda)
                indice = hilera_fila.find(palabra[::-1])
                if indice != -1:
                    posicion_inicio = (fila, indice + len(palabra) - 1)
                    posicion_final = (fila, indice)
                    direccion = '‚Üê'
                    encontrada = True

    return posicion_inicio, posicion_final, direccion
````

````python
def encontrar_palabra_vertical(sopa: list, palabra: str) -> tuple:
    """
    Busca una palabra en la sopa de letras de forma vertical.

    Args:
        sopa (list): Matriz que representa la sopa de letras.
        palabra (str): Palabra a buscar en la sopa de letras.

    Returns:
        tuple: Una tupla que contiene la posici√≥n de inicio (fila, columna),
               la posici√≥n final (fila, columna) y la direcci√≥n ('‚Üì' o '‚Üë').
               Si no se encuentra la palabra, devuelve (None, None, None).
    """
    # Ac√° su c√≥digo
    # PISTA: Puede usar la funci√≥n encontrar_palabra_horizontal como gu√≠a. Piense en c√≥mo recorrer la matriz por columnas en vez de filas.
    
def encontrar_palabra(sopa: list, palabra: str) -> tuple:
    """
    Busca una palabra en la sopa de letras en todas las direcciones (horizontal, vertical).

    Args:
        sopa (list): Matriz que representa la sopa de letras.
        palabra (str): Palabra a buscar en la sopa de letras.

    Returns:
        tuple: Una tupla que contiene la posici√≥n de inicio (fila, columna),
               la posici√≥n final (fila, columna) y la direcci√≥n ('‚Üí', '‚Üê', '‚Üì', '‚Üë').
               Si no se encuentra la palabra, devuelve (None, None, None).
    """
    # Buscar la palabra de forma horizontal
    posicion_inicio, posicion_final, direccion = encontrar_palabra_horizontal(sopa, palabra)
    # Si no se encontr√≥, buscar de forma vertical
    if posicion_inicio is None:
        posicion_inicio, posicion_final, direccion = encontrar_palabra_vertical(sopa, palabra)
    
    return posicion_inicio, posicion_final, direccion
````

````python
palabra = input("Ingrese la palabra a buscar: ").upper()

# Buscar la palabra en la sopa de letras
posicion_inicio, posicion_final, direccion = encontrar_palabra(sopa, palabra)

# Mostrar el resultado
if posicion_inicio is not None:
    print(f"Palabra encontrada de {posicion_inicio} a {posicion_final} en direcci√≥n {direccion}")
else:
    print("Palabra no encontrada")
````
