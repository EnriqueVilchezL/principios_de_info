## ¬øQu√© son los ejes (axes) en NumPy?

Los **ejes** son las dimensiones de un array de NumPy. Entender los ejes es fundamental para trabajar eficientemente con arrays multidimensionales.

### Regla fundamental:
**El eje (`axis`) especifica la dimensi√≥n A LO LARGO DE LA CUAL se realiza la operaci√≥n.**

Cuando usas `axis=0`, la operaci√≥n se realiza "bajando" por las filas (colapsa las filas).
Cuando usas `axis=1`, la operaci√≥n se realiza "a trav√©s" de las columnas (colapsa las columnas).

### Ejemplo pr√°ctico: Comprendiendo los ejes

Veamos c√≥mo funcionan las operaciones en diferentes ejes con ejemplos concretos:

````python
# Ejemplo 1: Array 2D - Operaciones con axis=0 y axis=1
print("=" * 60)
print("EJEMPLO 1: Array 2D")
print("=" * 60)

# Creamos un array 2D simple
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])

print("\nArray original:")
print(arr_2d)
print(f"Shape: {arr_2d.shape}  (2 filas, 3 columnas)")
print()

# Suma a lo largo del eje 0 (colapsa las filas - suma verticalmente)
suma_eje_0 = np.sum(arr_2d, axis=0)
print("np.sum(arr_2d, axis=0) - Suma a lo largo del eje 0:")
print(f"  [1+4, 2+5, 3+6] = {suma_eje_0}")
print(f"  Shape: {suma_eje_0.shape}  (solo quedan las columnas)")
print()

# Suma a lo largo del eje 1 (colapsa las columnas - suma horizontalmente)
suma_eje_1 = np.sum(arr_2d, axis=1)
print("np.sum(arr_2d, axis=1) - Suma a lo largo del eje 1:")
print(f"  [1+2+3, 4+5+6] = {suma_eje_1}")
print(f"  Shape: {suma_eje_1.shape}  (solo quedan las filas)")
print()

# Media a lo largo de diferentes ejes
media_eje_0 = np.mean(arr_2d, axis=0)
media_eje_1 = np.mean(arr_2d, axis=1)
print(f"np.mean(arr_2d, axis=0) = {media_eje_0}  (media de cada columna)")
print(f"np.mean(arr_2d, axis=1) = {media_eje_1}  (media de cada fila)")
print()

# M√°ximo a lo largo de diferentes ejes
max_eje_0 = np.max(arr_2d, axis=0)
max_eje_1 = np.max(arr_2d, axis=1)
print(f"np.max(arr_2d, axis=0) = {max_eje_0}  (m√°ximo de cada columna)")
print(f"np.max(arr_2d, axis=1) = {max_eje_1}  (m√°ximo de cada fila)")
````

### Resumen de conceptos clave:

| Operaci√≥n | axis=0 | axis=1 | axis=2 | axis=3 |
|-----------|--------|--------|--------|--------|
| **Array 2D** | Colapsa filas (‚Üì) | Colapsa columnas (‚Üí) | - | - |
| **Array 3D** | Colapsa capas (profundidad) | Colapsa filas | Colapsa columnas | - |
| **Array 4D** | Colapsa dim 0 | Colapsa dim 1 | Colapsa dim 2 | Colapsa dim 3 |

**Regla mnemot√©cnica:** 
- El n√∫mero del eje indica **cu√°l dimensi√≥n desaparece** del shape resultante
- Si `arr.shape = (2, 3, 4)` y aplicas `axis=1`, el resultado tiene shape `(2, 4)` - ¬°la dimensi√≥n 1 desapareci√≥!

## Ejercicios de Broadcasting üì°

### ¬øQu√© es Broadcasting?

**Broadcasting** es la capacidad de NumPy de realizar operaciones entre arrays de diferentes shapes autom√°ticamente, sin necesidad de copiar datos expl√≠citamente.

**Reglas de Broadcasting:**
1. Si los arrays tienen diferente n√∫mero de dimensiones, el de menor dimensi√≥n se "extiende" con 1s a la izquierda
2. Los arrays son compatibles en una dimensi√≥n si:
   - Tienen el mismo tama√±o en esa dimensi√≥n, O
   - Uno de ellos tiene tama√±o 1 en esa dimensi√≥n
3. El resultado tiene el tama√±o m√°ximo en cada dimensi√≥n

**Ejemplos visuales:**

```python
# Ejemplo 1: Vector + Escalar
arr = np.array([1, 2, 3])     # shape (3,)
escalar = 10                   # shape ()
resultado = arr + escalar      # shape (3,) ‚Üí [11, 12, 13]

# Ejemplo 2: Matriz + Vector (broadcasting en columnas)
matriz = np.array([[1, 2, 3],  # shape (2, 3)
                   [4, 5, 6]])
vector = np.array([10, 20, 30]) # shape (3,)
resultado = matriz + vector     # shape (2, 3)
# [[1+10, 2+20, 3+30],
#  [4+10, 5+20, 6+30]]

# Ejemplo 3: Matriz + Vector columna (broadcasting en filas)
matriz = np.array([[1, 2, 3],  # shape (2, 3)
                   [4, 5, 6]])
vector_col = np.array([[10],   # shape (2, 1)
                       [20]])
resultado = matriz + vector_col # shape (2, 3)
# [[1+10, 2+10, 3+10],
#  [4+20, 5+20, 6+20]]
```

### Funci√≥n Principal: Generar Reporte del Bosque üìä

Funci√≥n que combina todas las anteriores para generar un reporte completo.

````python
def generar_reporte_bosque(datos: np.ndarray, rangos_especies: list, nombres_especies: list) -> dict:
    """
    Genera un reporte completo del an√°lisis del bosque.
    
    Par√°metros:
    -----------
    datos : np.ndarray
        Array 3D con shape (meses, arboles, mediciones)
    rangos_especies : list
        Lista de tuplas [(inicio, fin)] para cada especie
    nombres_especies : list
        Lista con nombres de las especies
    
    Retorna:
    --------
    dict
        Diccionario con todos los resultados del an√°lisis
    
    Conceptos:
    ----------
    - Composici√≥n de funciones
    - Estructuraci√≥n de resultados
    - Integraci√≥n de m√∫ltiples an√°lisis
    """
    reporte = {
        'crecimiento': calcular_crecimiento(datos),
        'biomasa_especies': calcular_biomasa_por_especie(datos, rangos_especies),
        'arboles_alto_rendimiento': identificar_arboles_alto_rendimiento(datos),
        'indice_salud': calcular_indice_salud(datos),
        'nombres_especies': nombres_especies
    }
    return reporte
````
